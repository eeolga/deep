```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
import torch # Import torch

# Step 1: Read the Excel
file_path = '/content/logs_CAD_2024_1.xlsx'
def read_excel(file_path):
    df = pd.read_excel(file_path)
    return df

# Step 2: Select specific columns (second and fifth columns)
def select_columns(df):
    df_selected = df.iloc[:, [1, 3]]
    return df_selected

# Step 3: Convert to PyTorch tensor
def df_to_tensor(df):
    data_tensor = torch.tensor(df.values, dtype=torch.float32)
    return data_tensor

# Function to perform K-means clustering and calculate centroid distances
def kmeans_and_centroid_distances(data, num_clusters):
    # Handle missing values (NaN) before fitting KMeans
    data = data.dropna()  # Drop rows with missing values

    kmeans = KMeans(n_clusters=num_clusters, random_state=42, n_init=10)
    kmeans.fit(data)

    centroids = kmeans.cluster_centers_
    labels = kmeans.labels_

    # Calculate distances between centroids
    distances = np.linalg.norm(centroids[:, np.newaxis] - centroids, axis=2)

    # Add cluster labels to the dataframe
    data['Cluster'] = labels

    return centroids, distances, data

# Read the Excel file outside the main function
df = read_excel(file_path)

# Perform K-means clustering for 3 to 6 clusters
for n_clusters in range(3, 7):
    # Use the df DataFrame that was read outside the main function
    centroids, distances, clustered_data = kmeans_and_centroid_distances(df[['user_id', 'log_count']].copy(), n_clusters)

    print(f"Results for {n_clusters} clusters:")
    print("Centroids:\n", centroids)
    print("Centroid Distances:\n", distances)

    # Visualize the clusters
    plt.figure(figsize=(10, 6))
    sns.scatterplot(data=clustered_data, x='user_id', y='log_count', hue='Cluster', palette='viridis', s=100)
    plt.scatter(centroids[:, 0], centroids[:, 1], s=300, c='red', marker='X')
    plt.title(f'K-means Clustering with {n_clusters} Clusters')
    plt.xlabel('user_id')
    plt.ylabel('log_count')
    plt.legend(title='Cluster')
    plt.show()

# Main function to perform the steps (This function is not being used currently)
def main(file_path, num_clusters):
    df = read_excel(file_path)
    df_selected = select_columns(df)
    data_tensor = df_to_tensor(df_selected)
    # The kmeans_clustering function is not defined, you may need to implement it
    centroids, cluster_assignments = kmeans_clustering(data_tensor, num_clusters)
    # The plot_clusters function is not defined, you may need to implement it
    plot_clusters(data_tensor, cluster_assignments, centroids)
```
